```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
import pickle
import matplotlib.pyplot as plt
```

```python
with open('sam_big_results.pkl', 'rb') as f:
    arr = pickle.load(f)
    test_accuracy_history_sam = arr["acc"]
    test_loss_history_sam = arr["loss"]
    loss_in_training_sam = arr["loss_train"]
    time_learning_sam = arr["time"]
```

```python
colors = ['red', 'orange', 'deeppink', 'violet', 'mediumaquamarine', 'yellow', 'lime', 'gold', 'teal', 'royalblue', 'cyan', 'purple']
```

```python
plt.rc('xtick', labelsize=23)
plt.rc('ytick', labelsize=23)
```

```python
fig = plt.figure(figsize = (15, 10))
# plt.title('      ', fontsize=18)
plt.title(r'     $\alpha$  $\lambda$  ', fontsize=30)
plt.xlabel(' ', fontsize=27)
plt.ylabel(' ', fontsize=27)
j = 0
for param, accuracy in test_accuracy_history_sam.items():
    plt.plot(range(1, 21), accuracy, label = str(param), linewidth=3, color=colors[j])
    j += 1
plt.legend(fontsize = 25)
plt.grid(True)

plt.savefig('Test_accuracy_history_sam.pdf')
```

```python
fig = plt.figure(figsize = (15, 10))
# plt.title('          ', fontsize=18)
plt.title(r'      $\alpha$  $\lambda$  ', fontsize=30)
plt.xlabel(' ', fontsize=27)
plt.ylabel('  ', fontsize=27)
j = 0
for param, loss in test_loss_history_sam.items():
    plt.plot(range(1, 21), list(map(lambda x: x.item(), loss)), label = str(param), linewidth=3, color=colors[j])
    j += 1
plt.legend(fontsize = 25, loc='upper right')
plt.grid(True)

plt.savefig('Test_loss_history_sam.pdf')
```

```python
fig = plt.figure(figsize = (15, 10))
# plt.title('          ', fontsize=18)
plt.title(r'      $\alpha$  $\lambda$  ', fontsize=30)
plt.xlabel(' ', fontsize=27)
plt.ylabel('  ', fontsize=27)
j = 0
for param, loss in loss_in_training_sam.items():
    plt.plot(range(1, 21), loss, label = str(param), linewidth=3, color=colors[j])
    j += 1
plt.legend(fontsize = 25)
plt.grid(True)

plt.savefig('Train_loss_history_sam.pdf')
```
